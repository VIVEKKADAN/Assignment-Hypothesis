Null_Hypothesis=There is no significant difference between cutlet samples from two outlet: Mu1=Mu2

Alternate_Hypothesis=There is significant difference between cutlet samples from two outlet: Mu1!=Mu2

Alpha1=5%=0.05

```python
import pandas as pd
import scipy.stats as stats
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
Cutlets=pd.read_csv('Cutlets.csv')
Cutlets.head(5)
```

```python
Alpha1=0.05
T_test_Value=stats.ttest_ind(a=Cutlets['Unit A'],b=Cutlets['Unit B'])
```

```python
P1=T_test_Value[1]
print(P1)
if P1<Alpha1:
  print("Reject Null Hypothesis:There is significant difference between cutlet samples from two outlets")
else:
  print("Failed to Reject Null  Hypothesis:There is no significant difference between cutlet samples from two outlets")
```

```python
#Prforming ANOVA TESTING
Alpha=0.05
Anova=stats.f_oneway(LabTAT.iloc[:,0],LabTAT.iloc[:,1],LabTAT.iloc[:,2],LabTAT.iloc[:,3])
Anova
```

```python
Alpha=0.05
P=1-Alpha
chi_critic=stats.chi2.ppf(P,DOF)
print("Chi_square critical value is:",chi_critic)
#Checking the Hypothesis condition by using Cricritical and chi stats value:
if chi_critic<=STATS:
  print("Reject Null_Hypothesis: So the variables are dependent")
else:
  print("Failed to Reject Null_Hypothesis: So the variables are independent")
```

```python
#Checking the Hypothesis condition by using Percentage value with significance level:
if P<Alpha:
  print("Reject Null_Hypothesis: So the variables are dependent")
else:
  print("Failed to Reject Null_Hypothesis: So the variables are independent")
```
