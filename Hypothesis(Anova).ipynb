```python
# 25 Jan 2023
from scipy import stats
import pandas as pd
from scipy.stats import f_oneway
```

```python
#Null Hypothesis:All the 3 diet plans are the same
#Alternate Hypothesis:Alternate one of the diet plans is signifivcantly different
#from os import F_TEST
The_muppet_show=([11,7,8,14,11,10,5])
Futurama=[4,8,6,11,9,8]
BBC_NEWS=[4,3,2,2,3,6]
f_oneway(The_muppet_show,Futurama,BBC_NEWS)

```

```python
import pandas as pd
from scipy import stats
Atkins=pd.Series([6,2,3,4,2,3,3,2,7,8,10,4])
GM=pd.Series([4,4,5,7,8,5,3,7,10,4,5,2])
SouthBeach=pd.Series([6,7,5,6,8,7,8,9,6,5,5,12])
stats.f_oneway(Atkins,GM,SouthBeach)
```
